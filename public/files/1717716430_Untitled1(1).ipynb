```python
import networkx as nx
import itertools
import random

def gnp_random_graph(N, p):
    G = nx.Graph()
    G.add_nodes_from(range(N))

    for i, j in itertools.combinations(G.nodes, 2):
        r = random.random()
        if r < p:
            G.add_edge(i, j)

    return G
    G = gnp_random_graph(13, 0.17)
nx.draw(G)
print('Graph has', G.number_of_edges(), 'edges.')
```
